```python
data.dropna(subset=["Embarked"],inplace=True)
data["Cabin"].fillna("Unknown",inplace=True)
data["Age"].fillna(data["Age"].mean(),inplace=True)
```

Exploratory Data Analysis(EDA)

Survival Rate

```python
plt.figure(figsize=(8,4))
sns.countplot(data=data,x="Survived")
plt.title("Survival Count")
plt.xlabel("Survived")
plt.ylabel("Count")
plt.show()
```

Passenger Class Distribution

```python
plt.figure(figsize=(8,4))
sns.histplot(data['Pclass'],kde=True)
plt.title("Passenger Class Distribution")
plt.xlabel("Pclass")
plt.ylabel("Count")
plt.show()
```

Age Distribution

```python
plt.figure(figsize=(8,4))
sns.histplot(data["Age"],kde=True)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

Gender Distribution

```python
plt.figure(figsize=(8,4))
sns.countplot(data=data,x="Sex")
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

Embarked Port Distribution

```python
plt.figure(figsize=(8,4))
sns.histplot(data["Embarked"],kde=True)
plt.title("Embarked Port Distribution")
plt.xlabel("Embarked")
plt.ylabel("Count")
```

Survival Rate By Passenger Class

```python
plt.figure(figsize=(8,4))
sns.countplot(data=data,x="Pclass",hue="Survived")
plt.title("Survival Rate By Passenger Class")
plt.xlabel("Pclass")
plt.ylabel("Count")
plt.show()
```

Survival Rate By Gender

```python
plt.figure(figsize=(8,4))
sns.countplot(data=data,x="Sex",hue="Survived")
plt.title("Survival Rate By Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

Survival Rate By Age

```python
plt.figure(figsize=(8,4))
sns.histplot(data[data['Survived'] == 1]['Age'], bins=20, kde=True, color='Blue',label="Survived") # Change 'Label' to 'label'
sns.histplot(data[data['Survived'] == 0]['Age'], bins=20, kde=True, color='red',label="Not Survived") # Change 'Label' to 'label'
plt.title("Survival Rate By Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

Scatter plot Of Age Vs Fare

```python
plt.figure(figsize=(8,4))
sns.scatterplot(data=data,x="Age",y="Fare", hue="Age")
plt.title("Scatter plot of Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.legend(title="Survived")
plt.show()
```

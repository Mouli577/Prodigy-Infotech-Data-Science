```python
years = data.columns[1:]
total_values = data[years].sum()
plt.figure(figsize=(30, 30))
plt.barh(years, total_values, color = 'Red')
plt.xlabel('Total Values')
plt.ylabel('Year', size = 20)
plt.title('Total Values by Year', size= 20)
plt.show()
```
